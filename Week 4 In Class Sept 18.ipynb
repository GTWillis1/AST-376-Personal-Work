```python
from astropy.io import fits
from matplotlib import pyplot as plt
import numpy as np
import glob
%matplotlib inline

hdrList = []
im1 = fits.open('Sept18/2018_AST376_cals_0010.fits')
hdrList.append(im1[0].header)
im2 = fits.open('Sept18/2018_AST376_cals_0030.fits')
hdrList.append(im2[0].header)
im3 = fits.open('Sept18/2018_AST376_cals_0043.fits')
hdrList.append(im3[0].header)
im4 = fits.open('Sept18/2018_AST376_cals_0059.fits')
hdrList.append(im4[0].header)
im5 = fits.open('Sept18/2018_AST376_cals_0060.fits')
hdrList.append(im5[0].header)
im6 = fits.open('Sept18/2018_AST376_cals_0075.fits')
hdrList.append(im6[0].header)
im7 = fits.open('Sept18/2018_AST376_cals_0076.fits')
hdrList.append(im7[0].header)
im8 = fits.open('Sept18/2018_AST376_cals_0077.fits')
hdrList.append(im8[0].header)
      
data_list = []
im1_data = im1[0].data
data_list.append(im1_data)
im2_data = im2[0].data
data_list.append(im2_data)
im3_data = im3[0].data
data_list.append(im3_data)
im4_data = im4[0].data
data_list.append(im4_data)
im5_data = im5[0].data
data_list.append(im5_data)
im6_data = im6[0].data
data_list.append(im6_data)
im7_data = im7[0].data
data_list.append(im7_data)
im8_data = im8[0].data
data_list.append(im8_data)

for i in range(8):
    
    print('Image type: ', hdrList[i]['imagetyp'])
    print('Instrument filter: ', hdrList[i]['insfilte'])
    print('Exposure time: ', hdrList[i]['exptime'], '\n')
    
    print('Min: ',np.min(data_list[i]))
    print('Max: ',np.max(data_list[i]))
    print('Median: ',np.median(data_list[i]))
    print('Mean: ',np.mean(data_list[i]))
    print('Stdev: ',np.std(data_list[i]))
    histogram = plt.hist(data_list[i].flatten(), 100)
    plt.show()
    
filename = glob.glob('Sept18/2018_AST376_cals_007*.fits')
images = [ fits.getdata(image) for image in filename]
final_image = np.median(images, axis=0)
print('Min: ', np.min(final_image))
print('Max: ', np.max(final_image))
print('Median: ', np.median(final_image))
print('Mean: ', np.mean(final_image))
print('Stdev: ', np.std(final_image))

plt.figure()
vmin = np.median(final_image) - np.std(final_image)
vmax = np.median(final_image) + np.std(final_image)
plt.imshow(final_image, vmin=vmin, vmax=vmax)
plt.colorbar()
plt.tight_layout(); plt.show(); plt.close()
```
